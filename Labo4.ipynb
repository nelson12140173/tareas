```python
# 5) Tengo valores de cordenadas (x, y) en las columnas del array
# 1.33 4.5
# 30.0 10.7
# 70.2 0.5
# Agregue a este array las coordenadas (37.1, -3.6). Muestre en pantalla las dimensiones del nuevo array.
```

```python
import numpy as np

def agregar_coordenada(arr_origi, new_coord):
    """
    Agrega una nueva coordenada (x, y) al array original y muestra el nuevo array y sus dimensiones.
    
    Validaciones incluidas:
    - Que el array original sea de forma (n, 2)
    - Que la nueva coordenada tenga exactamente 2 elementos (x, y)
    
    Parámetros:
    - array_original: np.array con forma (n, 2)
    - nueva_coord: lista o array con una coordenada [x, y]
    
    Retorna:
    - El nuevo array con la coordenada agregada
    """
    # Validar que el array original tenga 2 columnas
    if arr_origi.ndim != 2 or arr_origi.shape[1] != 2:
        raise ValueError("debe tener exactamente 2 columnas (x, y).")
    
    # Validar que la nueva coordenada tenga 2 elementos
    if not isinstance(new_coord, (list, tuple, np.ndarray)) or len(new_coord) != 2:
        raise ValueError("La coordenada debe tener exactamente 2 elementos: [x, y].")
    
    new_coord = np.array([new_coord])  # Forma (1, 2)
    new_array = np.vstack([arr_origi, new_coord])
    
    print("Nuevo array con coordenadas:")
    print(new_array)
    print("Dimensiones del nuevo array:", new_array.shape)
    return new_array

coordenadas = np.array([
    [1.33, 4.5],
    [30.0, 10.7],
    [70.2, 0.5]
])

nueva = [37.1, -3.6]

agregar_coordenada(coordenadas, nueva)
```

```python
second_ciudad = poblacion_ordenada.sort_values(ascending=False).index[1]
print("La segunda ciudad más poblada es:", second_ciudad)
```

```python
# 6) Realice un histograma con la población de los Municipios para cada una de las provincias.
```

```python
import matplotlib.pyplot as plt

# se agrupa por provincia y municipio, y se suma la población total
poblacion = df.groupby(['Provincia', 'Municipio'])['Poblacion'].sum().reset_index()

# lista única de provincias
provincias = poblacion['Provincia'].unique()

# se crea un histograma por provincia
for provincia in provincias:
    datos_x_provincia = poblacion[poblacion['Provincia'] == provincia]['Poblacion']
    
    plt.figure(figsize=(8, 5))
    plt.hist(datos_x_provincia, bins=10, color='skyblue', edgecolor='black')
    plt.title(f"Histograma de población - {provincia}")
    plt.xlabel("Población de municipios")
    plt.ylabel("Frecuencia")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

```python
# 7) Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.
```

```python
import pandas as pd
import numpy as np

# se agrupa por municipio y se suma la población total
municipios = df.groupby('Municipio')['Poblacion'].sum().reset_index()

# se selecciona los 50 municipios aletorios
muestra = municipios.sample(n=50, random_state=42)  # random_state fija la selección

# calculo de estadísticas
promedio = muestra['Poblacion'].mean()
desviacion = muestra['Poblacion'].std()

# se muestra los Municipios seleccionados
print("Municipios seleccionados (con su población):")
print(muestra.sort_values(by='Poblacion', ascending=False).to_string(index=False))

print("\n Estadísticas de los 50 municipios aletorios:")
print("Promedio de población:", promedio)
print("Desviación estándar:", desviacion)
```

```python
# 8) Dígame los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.
```

```python
import numpy as np

# coordenadas de Madrid (X, Y)
coorde_madrid = df[df['Municipio'] == 'Madrid'][['X', 'Y']].iloc[0]

# Se excluye a Madrid
df_sin_madrid = df[df['Municipio'] != 'Madrid'].copy()

# calculo de distancia euclidiana a Madrid
df_sin_madrid['Distancia_a_Madrid'] = np.sqrt((df_sin_madrid['X'] - coorde_madrid['X'])**2 + 
                                               (df_sin_madrid['Y'] - coorde_madrid['Y'])**2)

# Encontrar el municipio más cercano y el más lejano
mas_cerca = df_sin_madrid.loc[df_sin_madrid['Distancia_a_Madrid'].idxmin()]
mas_lejos = df_sin_madrid.loc[df_sin_madrid['Distancia_a_Madrid'].idxmax()]

# Paso 5: Mostrar resultados
print("Municipio más cerca a Madrid:")
print("Nombre:", mas_cerca['Municipio'])
print("Provincia:", mas_cerca['Provincia'])
print("Distancia:", mas_cerca['Distancia_a_Madrid'])

print("\n Municipio más lejos a Madrid:")
print("Nombre:", mas_lejos['Municipio'])
print("Provincia:", mas_lejos['Provincia'])
print("Distancia:", mas_lejos['Distancia_a_Madrid'])
```
